### **Visualization with Plotly**

```python
# OHLC Chart
fig = go.Figure(go.Candlestick(x=data.index,
  open=data['Open'],
  high=data['High'],
  low=data['Low'],
  close=data['Close'], showlegend=False))

# Format
fig.update_layout(plot_bgcolor='rgba(0,0,0,0)',height=800,title="L'Occitane share price evolution", title_x=0.5)
fig.update_xaxes(showline=True, linewidth=2, linecolor='grey', gridcolor='lightgrey')
fig.update_yaxes(showline=True, linewidth=2, linecolor='grey', gridcolor='lightgrey')

# removing rangeslider
fig.update_layout(xaxis_rangeslider_visible=False)

# removing all empty dates
# build complete timeline from start date to end date
dt_all = pd.date_range(start=data.index[0],end=data.index[-1])
# retrieve the dates that ARE in the original datset
dt_obs = [d.strftime("%Y-%m-%d") for d in pd.to_datetime(data.index)]
# define dates with missing values
dt_breaks = [d for d in dt_all.strftime("%Y-%m-%d").tolist() if not d in dt_obs]
fig.update_xaxes(rangebreaks=[dict(values=dt_breaks)])

# add moving average to data
data['MA20'] = data['Close'].rolling(window=20).mean()

fig.add_trace(go.Scatter(x=data.index, 
                         y=data['MA20'], 
                         opacity=0.7, 
                         line=dict(color='orange', width=2), 
                         name='MA 20'))
```

```python
# MACD
macd = MACD(close=data['Close'], 
            window_slow=26,
            window_fast=12, 
            window_sign=9)
# stochastic
stoch = StochasticOscillator(high=data['High'],
                             close=data['Close'],
                             low=data['Low'],
                             window=14, 
                             smooth_window=3)

# add subplot properties when initializing fig variable
fig = make_subplots(rows=4, cols=1, shared_xaxes=True,
                    vertical_spacing=0.1, 
                    row_heights=[0.6,0.2,0.4,0.4],subplot_titles=("L'Occitane share price evolution", "L'Occitane volume evolution",
                                                                  "L'Occitane MACD evolution", "L'Occitane Stoch evolution"))

# Plot OHLC on 1st subplot (using the codes from before)
fig.add_trace(go.Candlestick(x=data.index,
  open=data['Open'],
  high=data['High'],
  low=data['Low'],
  close=data['Close'], showlegend=False), row=1, col=1 )

fig.add_trace(go.Scatter(x=data.index, 
                         y=data['MA20'], 
                         opacity=0.7, 
                         line=dict(color='orange', width=2), 
                         name='MA 20'))

# Plot volume trace on 2nd row
colors = ['green' if row['Open'] - row['Close'] >= 0 
          else 'red' for index, row in data.iterrows()]
fig.add_trace(go.Bar(x=data.index, 
                     y=data['Volume'],
                     marker_color=colors, marker_line_width = 0
                    ), row=2, col=1)

# Plot MACD trace on 3rd row
colors = ['green' if val >= 0 
          else 'red' for val in macd.macd_diff()]
fig.add_trace(go.Bar(x=data.index, 
                     y=macd.macd_diff(),
                     marker_color=colors, marker_line_width = 0
                    ), row=3, col=1)
fig.add_trace(go.Scatter(x=data.index,
                         y=macd.macd(),
                         line=dict(color='black', width=2)
                        ), row=3, col=1)
fig.add_trace(go.Scatter(x=data.index,
                         y=macd.macd_signal(),
                         line=dict(color='blue', width=1)
                        ), row=3, col=1)
# Plot stochastics trace on 4th row
fig.add_trace(go.Scatter(x=data.index,
                         y=stoch.stoch(),
                         line=dict(color='black', width=2)
                        ), row=4, col=1)
fig.add_trace(go.Scatter(x=data.index,
                         y=stoch.stoch_signal(),
                         line=dict(color='blue', width=1)
                        ), row=4, col=1)

# update layout by changing the plot size, hiding legends & rangeslider, and removing gaps between dates
fig.update_layout(height=900, width=1200, 
                  showlegend=False, 
                  xaxis_rangeslider_visible=False,
                  xaxis_rangebreaks=[dict(values=dt_breaks)],
                  plot_bgcolor='rgba(0,0,0,0)')
fig.update_xaxes(showline=True, linewidth=2, linecolor='grey', gridcolor='lightgrey')
fig.update_yaxes(showline=True, linewidth=2, linecolor='grey', gridcolor='lightgrey')
fig['layout']['yaxis2'].update(range=[0,10000000])

# update y-axis label
fig.update_yaxes(title_text="Price", row=1, col=1)
fig.update_yaxes(title_text="Volume", row=2, col=1)
fig.update_yaxes(title_text="MACD", showgrid=False, row=3, col=1)
fig.update_yaxes(title_text="Stoch", row=4, col=1)

fig.show()
```
